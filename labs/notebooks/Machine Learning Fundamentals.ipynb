# Regression using SGD:

```python
# Import libraries
import numpy as np

# Constants
n = 2  # number of inputs
m = 4  # number of training data samples

# Initialise SGD algorithm
theta = np.zeros((n, 1))
epsilon = 0.01

# Dummy input-output data
x = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
y = np.array([8, 7, 6, 5]).reshape(-1, 1)

# Loop over all training samples once (i.e. a single epoch)
for j in range(m):
    # Model prediction using a single data point
    yhat = np.dot(np.transpose(theta), x[:, j])

    # Estimate gradient of loss function
    g = (yhat - y[j]) * x[:, j]
    g = g[..., None]  # Force g to be column vector

    # SGD update rule
    theta = theta - epsilon * g
```
